

## Overview
::: {.justify}
Changes in oil prices have significant macroeconomic effects on countries and the global economy as a whole. The impact of oil price changes on macroeconomic variables depends on whether oil prices are rising or falling, and on the magnitude and persistence of the price change. Why does it have a significant impact? 

Central banks may also respond to inflationary pressures from higher oil prices by adjusting interest rates. Changes in oil prices can affect exchange rates. Oil-exporting countries may see their currencies strengthen with higher oil prices, while oil-importing countries may see their currencies weaken.

Higher oil prices can act as a drag on economic growth, particularly for countries that are net oil importers, potentially slowing down economic activity. Countries that are net oil exporters can benefit from rising oil prices by earning more revenue from oil exports, thereby improving their trade balance. 

There are voluntary,  explicit decisions that change the price of oil and, on occasionally, exogenous events that cause a significant change in the price of oil.

Geopolitical events and oil supply disruptions can lead to sudden and significant oil price spikes, which can have unpredictable effects on macroeconomic stability and global security.

Economic models and analysis are often used to assess and predict the potential macroeconomic volatility associated with oil price fluctuations.

It is important to develop a time series model based on historical data and events in order to predict possible future scenarios and thus anticipate the necessary measures to be implemented. This will lead to greater economic and social stability. 
:::

### Questions

1. How has the distribution of total global oil production among the main oil-producing countries evolved over the past two decades?
2. How do voluntary decisions made by organizations like OPEC (Organization of the Petroleum Exporting Countries), such as production cuts or increases, influence fluctuations in global oil prices?
3. How do exogenous factors and events influence the volatility and trends in global oil prices?
4. How have the US macroeconomic indicators, such as inflation rate, interest rates variation and currency exchange rate, responded to oil price shocks over the past years?
5. How have fluctuations in U.S. macroeconomic indicators influenced the global oil price over the past years?
6. How do fluctuations in oil prices impact stock market performance, such as energy, transportation, and staples sectors?
7. To what extent do variations in oil prices affect investor preferences for specific sectors, and how do they influence in stock market trends?
8. How do changes in oil prices affect consumers' attitudes and behaviors regarding personal finance, including spending and saving decisions?
9. How can time series analysis of historical oil price data be used to develop short-term and long-term forecasts of oil price fluctuations?
10. To what extent can the adoption of renewable energies and the reduction of exposure to fossil fuel price fluctuations contribute to improving economic stability?

## Big Picture
::: {.justify}
Petroleum and its derivatives are an essential raw material in almost every manufacturing process. Since there are currently no substitutes available in sufficient quantities, their price is a determining factor in the cost structure of the final products. Therefore, any significant change in price has a direct impact on the real economy. This forces governments to make decisions to mitigate potential negative effects by adjusting public economic policies.

When oil prices rise, consumers may reduce their spending on other goods and services due to higher energy costs. Consumer confidence may also be adversely affected, thereby affecting spending patterns.

Higher oil prices increase production costs for businesses, potentially reducing their profit margins. This may lead to reduced business investment, particularly in energy-intensive industries.

Rising oil prices may encourage investment in the energy sector, including exploration and production. Conversely, falling oil prices may lead to reduced investment in the energy sector, affecting employment and economic activity in related industries.
:::
***Big Picture Chart:***
![](./images/big_picture_diagram.png)


## Literature Review
::: {.justify}
A clear example of a voluntary decision is when OPEC (Organization of Petroleum Exporting Countries) decides to reduce the volume of production with the intention of causing an increase in the price of oil and its derivatives.

A geopolitical event, such as Russia's invasion of Ukraine, where a unilateral decision by Russia not to supply oil and gas to the member countries of the European Union, creates a potential shortage, causing the price of oil to rise.

The current situation continues to put inflationary pressure on the US economy, forcing the government to monitor the inflation rate, continue the process of raising the interest rates, in order to strengthen its currency and make oil imports cheaper. In this context, the U.S. and the European Union are obliged to accelerate the process of replacing the use of fossil fuels with renewable, environmentally friendly energy, thus generating less dependence on oil.
:::
